```python
from ydata_profiling import ProfileReport
profile = ProfileReport(df, title='Credit Card Dataset Profile Report')
profile
```

```python
for column in df.columns:
    if df[column].dtype == 'object':
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        df[column].fillna(df[column].mean(), inplace=True)

df
```

```python
# Extraction des données
data = df[['PURCHASES', 'CREDIT_LIMIT']]
data
```

```python
# Regroupement hiérarchique

from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='complete')
clust_labels=model.fit_predict(data)
```

```python
# Etiquétage

agglomerative=pd.DataFrame(clust_labels)

agglomerative
```

```python
# Clusters

import matplotlib.pyplot as plt

fig =plt.figure()

ax = fig.add_subplot(111)

scatter = ax.scatter (data ['PURCHASES'] , data ["CREDIT_LIMIT"] , c= agglomerative[0], s=50)

ax.set_title("Agglomerative Clutering")

ax.set_xlabel("Achats")

ax.set_ylabel("Limite de Crédit")

plt.colorbar(scatter)
```

```python
# Clustering Partitinnel

from sklearn.cluster import KMeans
Sum_of_squared_distances = []
K = range (1, 15)
for k in K :
  km = KMeans(n_clusters = k)
  km = km.fit(data)
  Sum_of_squared_distances.append(km.inertia_)
```

```python
# Visualisation

plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('La méthode de Elbow du k Optimal')
```

```python
# Notons k optimal = 3 (la méthode du coude)
```

```python
from sklearn.cluster import KMeans
Sum_of_squared_distances = []
K = range (1, 15)
for k in K :
  km = KMeans(n_clusters = 10)
  km = km.fit(data)
  Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sommes des carrées des distances')
plt.title('La méthode de Elbow du k Optimal')
```
